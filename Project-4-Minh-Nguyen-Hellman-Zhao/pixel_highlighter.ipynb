## Using this notebook

This notebook just uses the `plotly` library to display an image -- it highlights which pixel you are looking at. Look at the upper right corner for buttons to zoom in, pan, etc. Feel free to duplicate the block in order to more easily see two images simultaneously. Unfortunately with the size of our images, we cannot display two in the same block.

```
import cv2
import numpy as np
import plotly.express as px
```

```
############### Replace Image Filename ###############
filename = '/work/data/cube_basket.png'
#######

img1 = cv2.cvtColor(cv2.imread(filename), cv2.COLOR_BGR2RGB)

fig = px.imshow(img1)

fig.show()
```
